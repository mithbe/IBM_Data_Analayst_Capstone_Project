<h2>1. Load the Dataset</h2>

The functions below will download the dataset into your browser:



```python
# Define the URL of the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
## Write your code here
df.info()

df.describe()
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
plt.figure(figsize=(12,6))  # Set the figure size
sns.heatmap(df.isnull(), cmap='viridis', cbar=True, yticklabels=False)
plt.title("Heatmap of Missing Values")
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
missing_values_emp = df['Employment'].isnull().sum()
print("Missing Values for Employment:")
print(missing_values_emp)
```

```python
missing_values_comp = df['CompTotal'].isnull().sum()
print("Missing Values for CompTotal:")
print(missing_values_comp)
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
majority_value = df['CodingActivities'].mode()[0]
print(majority_value)
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
# Employment has 0 missing values so I will use CodingActivities

df['CodingActivities'].fillna(majority_value, inplace=True)
missing_values_cod = df['CodingActivities'].isnull().sum()
print("Missing Values for CodingActivities:")
print(missing_values_cod)
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here
# Get top 10 categories
top_10 = df['CodingActivities'].value_counts().nlargest(10).index

# Filter to top 10
df_top10 = df[df['CodingActivities'].isin(top_10)]

plt.figure(figsize=(12,6))
sns.countplot(data=df_top10, x='CodingActivities', order=top_10)

plt.title("Top 10 CodingActivities After Imputation", fontsize=16)
plt.xlabel("CodingActivities", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotate and align labels

plt.tight_layout()  # Adjust layout so labels fit nicely
plt.show()
```
